```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("/content/titanic.csv")
df.head()
```

```python
df.shape
```

```python
#checking for missing values
print("before preprocessing")
df.isnull().sum()
```

```python
#missing values are there in age, cabin and embarked column
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df.drop(columns=['Cabin'], inplace=True)  # Too many missing values
```

```python
# Encoding the categorical variables
df['Sex'] = df['Sex'].map({'male': 0, 'female': 1})

# Normalize data (optional)
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df[['Age', 'Fare']] = scaler.fit_transform(df[['Age', 'Fare']])
```

```python
df.isnull().sum()
```

```python
#Save the cleaned dataset
df.to_csv('cleaned_titanic.csv', index=False)

#Reloading the cleaned dataset
df_cleaned = pd.read_csv('cleaned_titanic.csv')
```

```python
#visualization
sns.boxplot(x='Survived', y='Fare', data=df_cleaned)
plt.title('Fare by Survival')
plt.show()
print("obs:The median fare for survivors is significantly higher than for non-survivors.")
```

```python
sns.histplot(df_cleaned['Age'], kde=True, bins=30)
plt.title('Age Distribution')
plt.xlabel('Normalized Age')
plt.show()
print("observation:The age distribution shows a peak around the normalized mid-range (~0.3–0.6), which corresponds to the real age range of 20–40 years and fewer passengers were in very young or very old age groups.")
```

```python
#scatterplot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df_cleaned, x='Age', y='Fare', hue='Survived', palette='coolwarm')
plt.title('Scatter Plot of Age vs Fare')
plt.xlabel('Normalized Age')
plt.ylabel('Normalized Fare')
plt.legend(title='Survived')
plt.show()
print("observation:Higher fare-paying passengers had a greater chance of survival")
```
